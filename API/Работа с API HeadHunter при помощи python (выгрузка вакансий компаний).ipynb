```python
# импорт библиотек
import requests
import json
import pandas as pd
```

```python
# Функция для выгрузки данных по вакансиям компаний с сайта hh.ru
def getPage(employer_id, area, page):
    params = {
        'employer_id': employer_id,      # ID компании на сайте hh.ru
        'area': area,                    # Поиск в зоне
        'page': page,                    # Номер страницы (максимум 20 страниц)
        'per_page': 100                  # Кол-во вакансий на 1 странице
    }   
    req = requests.get('https://api.hh.ru/vacancies', params)
    data = req.content.decode() 
    req.close()
    return data
```
